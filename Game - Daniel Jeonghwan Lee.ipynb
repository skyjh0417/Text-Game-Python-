```python
# Stage1 Game: Rock, Paper, Scissors

# Stage1 function that includes while loop to earn 3 wins
def stage1(name):
    print(f'''\nDear, {name}. This is yet the first challenge you will encounter!!!''')
    time.sleep(2)
    print(f'''\nThis is where 'Kawee Bawee Bo', the MASTER of Rock, Paper, Scissors is guarding against a challenger.\n''')
    time.sleep(2)
    print(f'''Fortunately, you only need to win 2 times whereas the master gets a handicap and needs to win 5 times.!!! \n''')
    time.sleep(2)
    print(f'''But, please be cautious for the master is the MASTER, indeed!!! \n''')
    time.sleep(2)
    print(f'''Are you able to withstand 'Kawee Bawee Bo', the MASTER of Rock, Paper, Scissors???!!!''')
    
    
    # These are the variables used in the stage
    RPS_list = ['Rock','Paper','Scissors']
    
    # new input will be added to the list once user inputs it
    player = []
    
    # random inputs will be assigned to the list
    computer = []
    
    # it counts how many times the user / computer wins
    player_won = 0
    computer_won = 0
    
    while True:         
        # A player gets to choose
        p_selection(RPS_list, player)
        
        time.sleep(1)
        # A computer gets to choose
        c_selection(RPS_list, computer)

        # Checking the result of the round
        time.sleep(1)
        if win(player, computer):
            player_won += 1
            print(f'''You just won the round. Congratulations. You won {player_won} time(s)!!!\n''')
        elif draw(player, computer):
            print(f'''It is a draw.\n ''')
        else:
            computer_won += 1
            print(f'''You just lost the round. Better luck next time! The master won {computer_won} time(s)!!!\n''')
        
        # Checking if player won 2 times. 
        time.sleep(1)
        if player_won == 2:
            print(f'''
            Congratulations! Dear, {name}!!!
            We are one step close to the escape!!!\n''')
            stage2(name)
            break
        
        # Checking if computer won 6 timres
        time.sleep(1)
        if computer_won == 6:
            print(f'''Oh... no... why did I ever trust you!!!\n''')
            gameover()
            break
    

# Append a value chosen by a player into a player's list
def p_selection(RPS_list, player):
    while True:
        time.sleep(1)
        RPS = input("""
    What would you choose? Type Rock, Paper, or Scissors or a number from 1 to 3.
    
    1) Rock
    2) Paper
    3) Scissors\n""")
        
        # if rock / paper/ siccsors / are typed
        time.sleep(1)
        if RPS.lower().strip() == 'rock':
            print(f'''\nYou selected {RPS_list[0]}\n''')
            player.append(RPS_list[0])
            break
            
        elif RPS.lower().strip() == 'paper':
            print(f'''\nYou selected {RPS_list[1]}\n''')
            player.append(RPS_list[1])
            break
            
        elif RPS.lower().strip() == 'scissors':
            print(f'''\nYou selected {RPS_list[2]}\n''')
            player.append(RPS_list[2])
            break
        
        # if a number is typed
        else:
            try:
                RPS_IND = int(RPS) -1
                if RPS_IND in range(0,3):
                    print(f'''\nYou selected {RPS_list[RPS_IND]}\n''')
                    player.append(RPS_list[RPS_IND])
                    break
                else:
                    print('Please type a number only from 1-3\n')
            except ValueError:
                print('Please type an interger from 1-3 only\n')

# Append a random value chosen by a computer into a computer's list
def c_selection(RPS_list, computer):
    computer_choice = RPS_list[random.randint(0,2)]    
    computer.append(computer_choice)
    print(f'''The master chose {computer_choice}\n''')

# Return True if a player wins the game
def win(player, computer):
    if ((player[-1] == 'Rock') & (computer[-1] == 'Scissors')) or ((player[-1] == 'Scissors') & (computer[-1] == 'Paper')) or ((player[-1] == 'Paper') & (computer[-1] == 'Rock')):
        return True
# Return True if the game is a draw
def draw(player, computer):
    if player[-1] == computer[-1]:
        return True
    
```

```python
# stage 2 Game: guessing a number
def stage2(name):
    print(f'''WOW!!! you made it, Dear {name}!!!!''')
    time.sleep(2)
    
    print(f'''\nYou are the third person who made to this challenge.''')
    time.sleep(2)
    
    print(f'''\nBut, please be aware, for there guards a gussing master, 'Soo Soo Kke Kki'!!!!''')
    time.sleep(2)
    
    print(f'''\nLike the first challenge, you are given 8 attempts!!! ''')
    time.sleep(2)
    
    print(f'''\nAnd, you need to guess a SECRET NUMBER between 1-100!!!''')
    time.sleep(2)
    
    print('\nPlease note! If your last guess is closer to the secret number than the previous guess,')
    time.sleep(2)
    
    print('\nit will say it\'s getting closer; otherwise, getting further')
    time.sleep(2)
    
    print(f'''\nPrevious guess: 8, Last guess: 30, Secret number: 27 - getting closer''')
    time.sleep(2)
    
    print(f'''\nPrevious guess: 8, Last guess: 30, Secret number: 15 - getting further''')
    time.sleep(2)
    
    # generaate_num() randomly chooses an int between 1-100 and assign it into secret_num
    secret_num = generate_num()
    
    # user is given 8 attempts
    guess_attempts = 8
    
    #user inputs are saved in the list
    guess_list = []
    
    #thiis the while loop that is valid only when the number of attemts is higher than 0
    while guess_attempts > 0:
        #this degrades the number of attempts by 1
        guess_attempts -= 1
        
        #it appends a user input to a guess_list
        guess_num(guess_list)
        
        time.sleep(1.5)   
        
        # if secret_num equals to the the last element of the list that user inputted
        if secret_num == guess_list[-1]:
            #it shows in how many attempts the user guessed it right
            print(f'''
            WOW!!! Dear {name}!!! You have guessed it right in {abs(guess_attempts-8)} attempt(s)!!!.
            Let's proceed to final challenge!!!
            ''')
            time.sleep(1.5)
            stage3(name)
            break
        
        # if guess_attmepts reaches to 0 it's a gameover
        elif guess_attempts == 0:
            print(f'''\nOh... no... You couldn't guess the secret number, {secret_num}...''')
            time.sleep(2)
            print(f'''\nWe will be stuck here forever... and die... alone... lonely...!!!''')
            
            gameover()
            break
        # for all other conditions, the user is given two hints. higher / lower and closer / farther
        else:
            high_low(secret_num, guess_list)
            time.sleep(1.5)
            distance(secret_num, guess_list)
            time.sleep(1.5)
            print(f'''
            You have {guess_attempts} attempt(s) left!!!.
            Think more carefully!!!
            ''')
    

# generate a random number from 1-100
def generate_num():
    return random.randint(1,100)

# ask a user to guess the number
def guess_num(guess_list):
    while True:
        try:
            answer = int(input('\nPlease input an integer from 1-100:\n'))
            # append the answer only if it is in the range of 1-100 and not in the guess_list
            if ((answer >=1) & (answer <= 100)) & (answer not in guess_list):
                guess_list.append(answer)
                break
            elif answer in guess_list:
                print('\nYou have already inputted the same number before. Please type in a new number for your own sake!')
            else:
                print('\nOut of range. Please input a number from 1-100')
        except:
            print('\nPlease intput an integer only')    

# tell if the guess number is higher or lower than the secret number
def high_low(secret_num, guess_list):
    if secret_num > guess_list[-1]:
        print(f'''\nThe secret number is higher than your guess, {guess_list[-1]}.''')
    else:
        print(f'''\nThe secret number is lower than Your guess, {guess_list[-1]}.''')

# if the difference between the guess_num and the secret num gets smaller, it tells getting closer and, otherwise, getting further.
def distance(secret_num, guess_list):
    try:
        if abs(secret_num - guess_list[-2]) > abs(secret_num - guess_list[-1]):
            print('\nYou are getting closer to the secret number.')
        elif abs(secret_num - guess_list[-2]) < abs(secret_num - guess_list[-1]):
            print('\nYou are getting further away from the secret number.')
    except IndexError:
        pass
    
```

```python
#stage 3 tic tac toe
def stage3(name):
    print(f'''WOW!!! Dear {name}!!!! You are the first person to make it to this level!!!!''')
    time.sleep(2)
    
    print(f'''\nI knew all along!!!! Your heroic movements can save me after all these hardships!!!!''')
    time.sleep(2)
    
    print(f'''\nThe last challenge is guarded by the king of the cave, the master 'Tic Tac Toe'!!!!''')
    time.sleep(2)
    
    print(f'''\nThe first person making a straight line by connecting 3 dots will win this challenge!!! ''')
    time.sleep(2)
    
    print(f'''\nAre you able to withstand the fury of the king, 'Tic Tac Toe'???!!!''')
    time.sleep(2)
    
    # game start
    game_on = True
    
    # a list to which a marker is placed
    b_list = [i for i in range(1,10)]
    
    # a list of marker
    marker = ['X','O']   
    
    #display the board
    board_display(b_list)
    time.sleep(1.5)
    
    #ask a player to choose the marker
    p_marker = choose_marker()
    # computer's marker is assigned
    c_marker = 'X' if p_marker == 'O' else 'O'
    
    time.sleep(1.5)
    # decide who goes first    
    turn = first_player(marker, p_marker)
    
    time.sleep(1.5)
    #while game_on is True, the while loop gets prompted
    while game_on:
        # there are two players. There fore player's switch should be on
        if turn == 'player':
            # choose a slot
            p_slot = p_choose_slot(b_list)
            
            time.sleep(1.5)
            # placing a marker
            place_marker(p_slot, p_marker, b_list)
            
            time.sleep(1.5)
            #display board
            board_display(b_list)
            
            #always checks the result every time the marker is placed on the board
            if tic_win(b_list):
                print(f'''\nDear {name}!!!! You have withstood all three challenges without fear!!!. Congratualations!''')
                time.sleep(2)
                
                print(f'''\nWe are finally free!!!!!!!!''')
                time.sleep(2)
                
                print('\nHope you enjoyed the game! :)')
                break
            
            elif tic_draw(b_list):
                while True:
                    response = input("It's a draw! Do you want to play the stage again? Press Y or N. ").upper().strip()
                    time.sleep(1.5)
                    if response == "Y" or response == "Yes":
                        stage3(name)
                        break
                    elif response == "N" or response == "No":
                        gameover()
                        game_on = False
                        
                    else:
                        print("Please type Y or N only!")                 
            #flip a turn to a computer
            turn = 'computer'
        
        # if turn equals to computer, this gets prompted
        else:
            #a slot is randomly chosen unless the the center slot is not filled
            c_slot = c_choose_slot(b_list)
            #places a marker
            place_marker(c_slot, c_marker, b_list)
            
            time.sleep(1.5)
            board_display(b_list)
            
            #always check the result
            if tic_win(b_list):
                print(f'''\nOh!!! No!!! You lost!!! I should have never followed you on this journey!!!''')
                time.sleep(2)

                print(f'''\nYou raised my hopes for escape and shattered them into pieces!!!''')
                time.sleep(2)

                print(f'''\nWhat a fool I was to ever trust you!!!''')
                time.sleep(2)
                
                gameover()
                game_on = False
                
            
            elif tic_draw(b_list):
                while True:
                    response = input("\nIt's a draw! Do you want to play the challenge again? Press Y or N. ").upper().strip()
                    if response == "Y" or response == "Yes":
                        stage3(name)
                        break
                    elif response == "N" or response == "No":
                        gameover()
                        game_on = False
                        
                    else:
                        print("\nPlease type Y or N only!")  
            # flip a turn
            turn = 'player'

# display the board
def board_display(b_list):

    print(f'''
    | {b_list[0]} | {b_list[1]} | {b_list[2]} |
    {'-'*13}
    | {b_list[3]} | {b_list[4]} | {b_list[5]} |
    {'-'*13}
    | {b_list[6]} | {b_list[7]} | {b_list[8]} |
    ''')
    
# ask a player to choose the marker
def choose_marker():
    while True:
        p_marker = input(f'''\nPlease choose a marker between 'X' and 'O' \n ''').upper().strip()
        if  p_marker == "X" or p_marker == "O":
            print(f'''\nYou selected {p_marker}! Good luck!!!''')
            return p_marker
            break
        else:
            print("\nYou didn't input correctly. Please type 'X' or 'O', again")
            
# check which marker goes first
def first_player(marker, p_marker):    
    turn = random.randint(0,1)
    if marker[turn] == p_marker:
        print("\nCongrats! You get to play first!!!")       
        return 'player'
        
    else:
        print("\nI'm sorry. The master goes first")
        return 'computer'

# ask where to put a marker
def p_choose_slot(b_list):
    # it only accepts int between 1-9 and also checks if the slot is filled or not. If filled, it asks again to select another slot
    while True:
        try:
            place = int(input('\nWhere do you want to place your marker? Type 1 - 9: \n'))
            if (place >=1) & (place <=9):
                if type(b_list[place - 1]) != str:
                    return place - 1
                    break
                else:
                    print('\nThe slot is filled. Please use a different slot')
            else:
                print('\nPlease type a number between 1-9 and not filled with a marker!')
        except:
            print('\nPlease type a number 1-9 only!')

# computer choose its slot
def c_choose_slot(b_list):
    #a list of an empty slot
    slot_not_filled = list(filter(lambda i: type(i) != str, b_list))
    
    # take the center slot first
    if 5 in slot_not_filled:
        return 4
    # if center slot is taken, choose a random slot among the slots not filled
    else:
        slot = random.choice(slot_not_filled) - 1
        return slot


# placing a marker into the list
def place_marker(slot, marker, b_list):
    b_list[slot] = marker
    
# check if win
def tic_win(b_list):
    if b_list[0] == b_list[1] == b_list[2] or b_list[3] == b_list[4] == b_list[5] or b_list[6] == b_list[7] == b_list[8] or b_list[0] == b_list[3] == b_list[6] or b_list[1] == b_list[4] == b_list[7] or b_list[2] == b_list[5] == b_list[8] or b_list[0] == b_list[4] == b_list[8] or b_list[6] == b_list[4] == b_list[2]:
        return True
    else:
        False

# check if the board is full
def tic_draw(b_list):
    return True if len(list(filter(lambda i: type(i) != str, b_list))) == 0 else False
    

```
